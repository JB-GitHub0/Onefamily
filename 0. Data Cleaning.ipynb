### Issues:
- Data mis-matches
- Missing values (left at things like 999 etc)
- Categorical varibales that may be ordinal

```python
def clean_df(df):
    """
    Replaces specific outlier/'fake-missings' values with NaN
    """

    # Dictionary mapping each column to a list of values to replace with NaN
    column_replacements = {
        "JF (Score)": [9999],
        "X (Months same person on ER at current address)": [-999997],
        "EF (No. of people not same surname at current address)": [-999997],
        "ND (Months since last CCJ)": [-999997],
        "NOB (Property group)": ["{ND}", "XX"],
        "URB (Income group)": ["{ND}", "X"],
        "LSB (Regional banded house price band)": [99, -999997],
        "BB (Number of CCJs)": [-999997],
    }

    # Apply replacements only to the specified columns
    for col, outlier_values in column_replacements.items():
        if col in df.columns:
            df[col].replace(outlier_values, np.nan, inplace=True)

    # Enforcing binary columns
    binary_feats = ["Smoker", "Joint?", "Sold"]
    yes_no_map = {"Yes": "Y", "No": "N"}
    binary_map = {"Y": 1, "N": 0}

    for feat in binary_feats:
        df[feat] = df[feat].map(yes_no_map).fillna(df[feat])
        df[feat] = df[feat].map(binary_map).fillna(df[feat])

    
    return df

df_cleaned = clean_df(df)
```

```python
# Run another report to check the changes have worked
CleanedReport = sv.analyze(df_cleaned)
CleanedReport.show_html("Sweetviz Reports\CleanedReport.html")
```

```python

# Save the clean df
df_cleaned.to_csv("Data\MockDataSet1_Cleaned.csv", index=False)
```
